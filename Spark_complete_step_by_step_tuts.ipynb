```
def count_nulls(df):
    null_counts = []          
    for col in df.columns:     
        
      nulls = df.where( df[col].isNull() ).count()
      result = tuple([col, nulls])  
      null_counts.append(result)      
    return (null_counts)
```

```
def imputer(df):
  num_cols=[i[0] for i in list(filter(lambda dtype: dtype[1] in {"float","int"},df.dtypes))]
  mean_dict = { col: 'mean' for col in num_cols }
  col_avgs = df.agg(mean_dict).collect()[0].asDict()
  mean_subs={k[4:-1]:v for k,v in col_avgs.items()}
  df=df.fillna(mean_subs)
  return(df)
```

```
# Now we will look at the application of udf to create new columns within a dataframe based on apllying a user defined function on an existing column
#examples of udf
# PySpark UDFs work in a similar way as the pandas .map() and .apply() methods for pandas series and dataframes. 
# If I have a function that can use values from a row in the dataframe as input, then I can map it to the entire dataframe. 
# The only difference is that with PySpark UDFs I have to specify the output data type.
```

```
# example data
df_pd = pd.DataFrame(
    data={'integers_cols': [1, 2, 3],
     'floats_cols': [-1.0, 0.5, 2.7],
     'integer_arrays_cols': [[1, 2], [3, 4, 5], [6, 7, 8, 9]]}
)
df = spark.createDataFrame(df_pd)
df.printSchema()
```

```
df.show()
```

```
# we will apply squaring function
def square(x):
  return(x**2)
```

```
# next step would be to convert this function into pyspark function and also declaring the outpouy type expected from the function
square_udf=udf(lambda x:square(x),IntegerType())
```

```
# now is the application part of the function on a specific column but using either .select() or .withColumns()
df.select("integers_cols","floats_cols",square_udf("integers_cols").alias("Squared_int_cols"),square_udf("floats_cols").alias("Squared_float_cols")).show()
```

```
# This is the diffrence between normal python and pyspark function - it will work only on that columsn whose datatypes match so integer funcs wont work in float and viceversa 
```

```
# To rectify the issue
# we will apply squaring function converting into float
def square(x):
  return(float(x**2))
```

```
# next step would be to convert this function into pyspark function and also declaring the output type as float
square_udf=udf(lambda x:square(x),FloatType())
```

```
# now is the application part of the function on a specific column but using either .select() or .withColumns()
df.select("integers_cols","floats_cols",square_udf("integers_cols").alias("Squared_int_cols"),square_udf("floats_cols").alias("Squared_float_cols")).show()
```

```
#If the input of the Python function is a list, then the output values will be the list having the output of the same datatype, 
#which is specified within ArrayType(FloatType()) when registering the UDF
```

```
def square_list(x):
    return [float(val)**2 for val in x]

square_list_udf = udf(lambda y: square_list(y), ArrayType(FloatType())) # floattype() for each element and array for thr list

df.select('integer_arrays_cols', square_list_udf('integer_arrays_cols')).show()
```

```
# If you are using any numpy function like np.mean,np.sum ,np.......... use it as usual but if using on a array type col then use .tolist() begore returning
def square(x):
  return(float(np.square(x**2)))

square_udf=udf(lambda x:square(x),FloatType())
df.select("integers_cols","floats_cols",square_udf("integers_cols").alias("Squared_int_cols"),square_udf("floats_cols").alias("Squared_float_cols")).show()
```

```
def square_array_right(x):
    return np.square(x).tolist()

spark_square_array_right = udf(square_array_right, ArrayType(IntegerType()))

df.withColumn('squared', spark_square_array_right('integer_arrays_cols')).show()
```

```
# create a new columns based on a function that takes multiple columns as your pararmeter
def filter_condition(c1,c2,c3):
  if c1=="no" and c2=="Yes" :
    return("ok")
  elif c1=="no" and c3=="Yes":
    return("ok")
  elif c1=="Yes" and c2=="Yes" and c3=="Yes":
    return("nice")
  else:
    return("rejected")
```

```
udf_filter_condition=udf(lambda x,y,z:filter_condition(x,y,z),StringType())  # or you can only mention the function name with
```

```
spark_df.withColumn("rating",udf_filter_condition("col1","col2","col3")).show()
```
